#Average Tip Amount by Each day of month

select avg(f.tip_amount) as Avg_tip_amount,dd.pick_day

from

gentle-presence-312618.uber_de_project.fact_table f

join

gentle-presence-312618.uber_de_project.datetime_dim dd

on f.datetime_id=dd.datetime_id

group by dd.pick_day

order by Avg_tip_amount desc

#Fare amount by payment type

select sum(f.fare_amount),pd.payment_type_name from

gentle-presence-312618.uber_de_project.fact_table f join
gentle-presence-312618.uber_de_project.payment_type_dim pd

on f.payment_type_id = pd.payment_type_id

group by pd.payment_type_name

#Number of trip made each day of week

Select count(\*) as Number_of_Trips,dd.pick_weekday

from

gentle-presence-312618.uber_de_project.fact_table f

join

gentle-presence-312618.uber_de_project.datetime_dim dd

on f.datetime_id=dd.datetime_id

group by dd.pick_weekday

#Top 5 longest trips

Select trip_id,tolls_amount,total_amount,td.trip_distance

from

gentle-presence-312618.uber_de_project.fact_table f

join

gentle-presence-312618.uber_de_project.trip_distance_dim td

on f.trip_distance_id=td.trip_distance_id

order by td.trip_distance desc Limit 5

#Total Revenue for each rate type

Select rd.rate_code_name,sum(total_amount) as Total_Revenue

from

gentle-presence-312618.uber_de_project.fact_table f

join

gentle-presence-312618.uber_de_project.rate_code_dim rd

on f.rate_code_id=rd.rate_code_id

group by rd.rate_code_name

#Trip and Fare metrics by ratecode and payment

SELECT rd.rate_code_name, pd.payment_type_name,

AVG(td.trip_distance) AS avg_trip_distance,

AVG(f.fare_amount) AS avg_fare_amount,

AVG(f.tip_amount) AS avg_tip_amount,

AVG(f.total_amount) AS avg_total_amount

FROM

gentle-presence-312618.uber_de_project.fact_table f

join

gentle-presence-312618.uber_de_project.payment_type_dim pd

on

f.payment_type_id=pd.payment_type_id

join

gentle-presence-312618.uber_de_project.rate_code_dim rd

on f.rate_code_id=rd.rate_code_id

join gentle-presence-312618.uber_de_project.trip_distance_dim td

on f.trip_distance_id=td.trip_distance_id

GROUP BY rd.rate_code_name, pd.payment_type_name;
